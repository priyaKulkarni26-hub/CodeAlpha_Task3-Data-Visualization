```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
print("Rows and Columns:", df.shape)
df.info()
df.isnull().sum()
df.drop_duplicates(inplace=True)
plt.figure(figsize=(6,4))
```

```python
sns.countplot(x="type", data=df)

plt.title("Movies vs TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
plt.figure(figsize=(10,5))

df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.show()
plt.figure(figsize=(12,5))
```

```python
df["release_year"].value_counts().sort_index().plot()

plt.title("Content Released by Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
```

```python
sns.countplot(
    y="rating",
    data=df,
    order=df["rating"].value_counts().index[:10]
)

plt.title("Top Ratings")

plt.show()
plt.figure(figsize=(10,6))
```

```python
df["listed_in"].value_counts().head(10).plot(kind="barh")

plt.title("Top Genres")

plt.xlabel("Count")

plt.show()
print("Conclusion")

print("- Movies are more than TV Shows.")
print("- Netflix content has increased over the years.")
print("- Most content comes from a few countries.")
print("- Different ratings are available for different age groups.")
print("- Data visualization helps understand trends easily.")
```
