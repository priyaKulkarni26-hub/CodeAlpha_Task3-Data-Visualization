# Task 3: Data Visualization
## Netflix Movies & TV Shows Analysis

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

df=pd.read_csv("netflix_titles.csv")
print(df.head())
print(df.shape)
print(df.isnull().sum())
df.drop_duplicates(inplace=True)

sns.countplot(x="type",data=df)
plt.title("Movies vs TV Shows")
plt.show()

df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries")
plt.show()

df["release_year"].value_counts().sort_index().plot(figsize=(10,4))
plt.title("Content Released by Year")
plt.show()

sns.countplot(y="rating",data=df,order=df["rating"].value_counts().index[:10])
plt.title("Top Ratings")
plt.show()

df["listed_in"].value_counts().head(10).plot(kind="barh")
plt.title("Top Genres")
plt.show()

print("Conclusion:")
print("- Movies are more than TV Shows")
print("- Content has increased over the years")
print("- Data visualization reveals useful insights")
```
